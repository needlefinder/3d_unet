```python
def browse_data(imgs,labels):
    n = imgs.shape[0]
    def view_image(sample):
        fig = plt.figure(figsize=(9, 9));
        plt.subplot(221)
        ax = plt.imshow(imgs[sample,0,:,:],cmap=plt.cm.gray, interpolation='none')
        fig.colorbar(ax)
        plt.subplot(222)
        ax = plt.imshow(labels[sample,0,:,:],cmap=plt.cm.gray)
        #fig.colorbar(ax)
        plt.subplot(223)
        if 'WG' not in uid:
            ax = plt.imshow(labels[sample,1,:,:],cmap=plt.cm.gray)
        #fig.colorbar(ax)
        plt.subplot(224)
        ax = plt.imshow(imgs[sample,0,92:-92,92:-92],cmap=plt.cm.gray)
        ax = plt.imshow(labels[sample,0,:,:],cmap=plt.cm.Reds,alpha=0.3)
        if 'WG' not in uid:
            ax = plt.imshow(labels[sample,1,:,:],cmap=plt.cm.Greens,alpha=0.3)
    interact(view_image, sample=(0,n-1))
```

```python
browse_data(imgs,labels)
```
